```python
import pandas as pd
train_data = pd.read_csv('./input/train_public.csv')
test_data = pd.read_csv('./input/test_public.csv')
train_data.head()
```

```python
train_data=train_data.fillna(method='ffill')
test_data=test_data.fillna(method='ffill')
```

```python
work_year_dict = {
    '< 1 year': 0,
    '1 year': 1,
    '2 years': 2,
    '3 years': 3,
    '4 years': 4,
    '5 years': 5,
    '6 years': 6,
    '7 years': 7,
    '8 years': 8,
    '9 years': 9,
    '10+ years': 10,
}

train_data['work_year'] = train_data['work_year'].map(work_year_dict)
test_data['work_year'] = test_data['work_year'].map(work_year_dict)
train_data.head()
```

```python
class_dict = {
    'A': 1,
    'B': 2,
    'C': 3,
    'D': 4,
    'E': 5,
    'F': 6,
    'G': 7,
}

train_data['class'] = train_data['class'].map(class_dict)
test_data['class'] = test_data['class'].map(class_dict)
train_data.head()
```

```python
train_data['issue_date'] = pd.to_datetime(train_data['issue_date'])
train_data['issue_date_month'] = train_data['issue_date'].dt.month
train_data['issue_date_dayofweek'] = train_data['issue_date'].dt.dayofweek

test_data['issue_date'] = pd.to_datetime(test_data['issue_date'])
test_data['issue_date_month'] = test_data['issue_date'].dt.month
test_data['issue_date_dayofweek'] = test_data['issue_date'].dt.dayofweek

train_data.head()
```

```python
#对'employer_type', 'industry'进行独热编码
from sklearn.preprocessing import OneHotEncoder
new1 = train_data[['employer_type','industry']]
new2 = test_data[['employer_type','industry']]
```

```python
enc1 = OneHotEncoder(categories = 'auto').fit(new1)
result1 = enc1.transform(new1).toarray()

enc2 = OneHotEncoder(categories = 'auto').fit(new2)
result2 = enc2.transform(new2).toarray()
```

```python
result1 = pd.DataFrame(result1)
result2 = pd.DataFrame(result2)

train_data = pd.concat([train_data,result1], axis = 1)
test_data = pd.concat([test_data,result2], axis = 1)

```

```python
#删除某些列
#'issue_date', 'earlies_credit_mon','employer_type', 'industry'
col_to_drop = ['issue_date', 'earlies_credit_mon','employer_type', 'industry']
train_data = train_data.drop(col_to_drop, axis=1)
test_data = test_data.drop(col_to_drop, axis=1)
train_data
```

```python
X = train_data.drop(columns='isDefault')
y = train_data['isDefault']

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2 ,random_state = 123)
```
